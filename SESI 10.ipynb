```python
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import numpy as np
import random
import warnings
import matplotlib.pyplot as plt
%matplotlib inline

sns.set(rc={'figure.figsize':(13,7.5)})
sns.set_context('talk')
```

```python
#generate data of people in pounds
np.random.seed(42)
normal_ditribution_us_male_mass_pounds=np.random.normal(loc=181,
                                                       scale=24,
                                                       size=6500)
normal_ditribution_us_female_mass_pounds=np.random.normal(loc=132,
                                                       scale=22,
                                                       size=6500)
all_mass_values=np.concatenate((normal_ditribution_us_male_mass_pounds,
                               normal_ditribution_us_female_mass_pounds),
                              axis=0)
df_ppl_mass=pd.DataFrame(data={'us_people_mass_pounds':all_mass_values})

df_ppl_mass.head()
```

```python
sns.distplot(df_ppl_mass['us_people_mass_pounds'],
            color='darkslategrey')
plt.xlabel('mass[pounds]', labelpad=14)
plt.ylabel('probability of occurance', labelpad=14)
plt.title('Distribution of Mass of People in U.S',
         y=1.015, fontsize=20)
```

```python
#calculation population mean
pop_mean_mass=df_ppl_mass['us_people_mass_pounds'].mean()
pop_mean_mass
```

```python
#calculate population standard deviation
pop_std_dev_mass=df_ppl_mass['us_people_mass_pounds'].std()
pop_std_dev_mass
```

```python
sample_means=[]
n=25
for sample in range(0,300):
    sample_values= np.random.choice(a=df_ppl_mass['us_people_mass_pounds'],
                                   size=n)
    sample_mean=np.mean(sample_values)
    sample_means.append(sample_mean)
```

```python
sns.distplot(sample_means)
plt.title("Distribution of Sample Means ($n=25$) of people's mass in pounds",
         y=1.015, fontsize=20)
plt.xlabel('sample mean mass[pounds]', labelpad=14)
plt.ylabel('frequency of occurence', labelpad=14)
```

```python
median_of_sample_means=np.median(sample_means)
median_of_sample_means
```

```python
mean_of_sample_means=np.mean(sample_means)
mean_of_sample_means
```

```python
pop_mean_mass
```

```python
std_dev_of_sample_means=np.std(sample_means)
std_dev_of_sample_means
```

```python
standard_error=pop_std_dev_mass/np.sqrt(n)
standard_error
```

```python
tail=1-stats.norm.cdf(1.96)
tail
```

```python
#area pada kurva diantara z-critical scores ada dikisaran 0.95
1-(tail*2)
```

```python
#batas nilai bawah untuk mean sample
lower_95_perc_bound= mean_of_sample_means-1.96*standard_error
lower_95_perc_bound
```

```python
upper_95_perc_bound=mean_of_sample_means+1.96*standard_error
upper_95_perc_bound
```

```python
kde=stats.gaussian_kde(sample_means)
pos=np.linspace(np.min(sample_means),
               np.max(sample_means), 10000)
plt.plot(pos, kde(pos), color='teal')
shade=np.linspace(lower_95_perc_bound,
                 upper_95_perc_bound,300)
plt.fill_between(shade, kde(shade), alpha=0.45,
                color='teal')
plt.text(x=156, y=.01, horizontalalignment='center',
        fontsize=17,
        s="95% of sample means fall withn\nthis shaded area of plus or minus \n1.96 z-scores from the mean",
        bbox=dict(facecolor='whitesmoke',
                 boxstyle='round, pad=0.1'))
plt.title("Distribution of Sample Means ($n=25$) of People's Mass in Pounds",
         y=1.015, fontsize=20)
plt.xlabel('Sample mean mass[pounds]', labelpad=14)
plt.ylabel('frequency of occurance', labelpad=14)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import scipy.stats.distributions as dist
df=pd.read_csv('heart.csv')
df.head()
```

```python
p_us=len(df[df['target']==1])/len(df)
p_us
```

```python
#standard error
se=np.sqrt(0.42*(1-0.42)/len(df))
#best estimate
be=p_us
#hypothesized estimate
he=0.42
test_stat=(be-he)/se
test_stat
```

```python
pvalue=2*dist.norm.cdf(-np.abs(test_stat))
pvalue
```
